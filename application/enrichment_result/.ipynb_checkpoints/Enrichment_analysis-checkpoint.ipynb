# Enrichment via gregor
## 1. Lift over, this is to lift over the snp into hg19 which gregor used

```bash
import cugg
import pandas as pd
from cugg import liftover

```

```bash
df = pd.read_csv("/home/hs3163/Enrichment_snp_postive_set.tsv",sep = "\t")
```

```bash
df
```

```bash
lf = liftover.Liftover('hg38','hg19')
snp_pos = df.apply(lambda x:  lf.chrpos_liftover(x["CHROM"],x["POS"]),axis = 1 )
df["POS_hg19"] =  [x[1] for x in snp_pos]
```

```bash
df.to_csv("/home/hs3163/Enrichment_snp_postive_set_lifted.tsv",sep = "\t")
```

```bash
h16 =  pd.read_csv("/mnt/vast/hpc/csg/molecular_phenotype_calling/QTL_fine_mapping/output/haQTL_pure_completed_unlimited_16_99pCS.tsv",sep = "\t")
m16 =  pd.read_csv("/mnt/vast/hpc/csg/molecular_phenotype_calling/QTL_fine_mapping/output/mQTL_pure_completed_unlimited_16_99pCS.tsv", sep = "\t")
```

```bash
df = df.assign(trait =lambda x: x['ID'].map(lambda id: 'ha' if id in h16['variants'].values else ('m' if id in m16['variants'].values )))
```

```bash
df[df.trait == "m"]
df[df.trait == "ha"]
```

```bash
df = df.assign(new_ID = df.CHROM.astype("str") + ":" + df.POS_hg19.astype("str") )
```

```bash
df.to_csv("/home/hs3163/Enrichment_snp_postive_set_lifted.tsv",sep = "\t")
```

```bash
for dir in */*/; do
    # Ensure that the directory is not empty
    if [[ -d "$dir" ]]; then
        # Calculate the total number of lines in neighbor.*.txt files
        total_lines=$(find "$dir" -name 'neighbor.*.txt' -exec cat {} + | wc -l)

        # Count the number of neighbor.*.txt files
        file_count=$(find "$dir" -name 'neighbor.*.txt' | wc -l)

        # Calculate the adjusted row number
        row_num=$((total_lines - file_count))

        # Check if PValue.txt exists
        if [[ -f "${dir}/PValue.txt" ]]; then
            # Append the row count to PValue.txt and save as PValue_new.txt
            (cat "${dir}/PValue.txt"; echo "N1_N2 = $row_num") > "${dir}/PValue_N1_N2.txt"
        fi
    fi
done
```

```bash
cd /mnt/vast/hpc/csg/molecular_phenotype_calling/QTL_fine_mapping/output/enrichment/ha/ 
# Loop through each subdirectory
for dir in */*/; do
    # Ensure that the directory is not empty
    if [[ -d "$dir" ]]; then
        # Calculate the total number of lines in neighbor.*.txt files
        total_lines=$(find "$dir" -name 'neighbor.*.txt' -exec cat {} + | wc -l)

        # Count the number of neighbor.*.txt files
        file_count=$(find "$dir" -name 'neighbor.*.txt' | wc -l)

        # Calculate the adjusted row number
        row_num=$((total_lines - file_count))

        # Check if PValue.txt exists
        if [[ -f "${dir}/PValue.txt" ]]; then
            # Append the row count to PValue.txt and save as PValue_new.txt
            (cat "${dir}/PValue.txt"; echo "N1_N2 = $row_num") > "${dir}/PValue_N1_N2.txt"
        fi
    fi
done

# Header for the new summary file
echo -e "Bed_File\tInBed_Index_SNP\tExpectNum_of_InBed_SNP\tPValue\tN1_N2\tNp\tNp_Nn" > /mnt/vast/hpc/csg/molecular_phenotype_calling/QTL_fine_mapping/output/enrichment/ha/Enrichment_snp_postive_set_lifted_ha_gregor_output_variant_counts.txt

# Calculate Np and Np_Nn for the neighbor_SNP directory
if [[ -d "*/neighbor_SNP" ]]; then
    if [[ -f "*/neighbor_SNP/index.snp.neighbors.txt" ]]; then
        Np=$(($(wc -l < "*/index_SNP/annotated.index.snp.txt") - 1))
    else
        Np=0
    fi
    Np_Nn_files=($(find "*/neighbor_SNP" -name 'neighbor.chr*txt'))
    Np_Nn=0
    for file in "${Np_Nn_files[@]}"; do
        Np_Nn=$(($Np_Nn + $(wc -l < "$file")))
    done
    Np_Nn=$(($Np_Nn - ${#Np_Nn_files[@]}))
fi

# Loop through each subdirectory
for dir in */*/; do
    # Ensure that the directory is not empty
    if [[ -d "$dir" ]]; then
        # Check if PValue_new.txt exists
        if [[ -f "${dir}PValue_N1_N2.txt" ]]; then
            # Extract values from PValue_new.txt
            inBedIndexSNPNum=$(grep "inBedIndexSNPNum" "${dir}PValue_N1_N2.txt" | cut -d '=' -f2 | tr -d '[:space:]')
            expectedS=$(grep "expectedS" "${dir}PValue_N1_N2.txt" | cut -d '=' -f2 | tr -d '[:space:]')
            p3=$(grep "p3" "${dir}PValue_N1_N2.txt" | cut -d '=' -f2 | tr -d '[:space:]')
            N1_N2=$(grep "N1_N2" "${dir}PValue_N1_N2.txt" | cut -d '=' -f2 | tr -d '[:space:]')
            # Get the directory name as the Bed_File
            bedFile=$(basename "$dir")
            # Append the data to the summary file
            echo -e "$bedFile\t$inBedIndexSNPNum\t$expectedS\t$p3\t$N1_N2\t$Np\t$Np_Nn" >> /mnt/vast/hpc/csg/molecular_phenotype_calling/QTL_fine_mapping/output/enrichment/ha/Enrichment_snp_postive_set_lifted_ha_gregor_output_variant_counts.txt
        fi
    fi
done
```
